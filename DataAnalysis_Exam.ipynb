**Построенная сводная таблица.**

```python
import pandas as pd

try:
    # Загрузка данных из файла
    file_path = "/content/sample_data/sales_data_large.csv"
    data = pd.read_csv(file_path)

    # Анализ данных
    total_records = data.shape[0]  # Общее количество записей

    # Уникальные регионы и продукты
    regions = data['region'].unique()
    products = data['product'].unique()

    # Общие продажи по регионам
    total_sales_by_region = data.groupby('region')['sales'].sum()

    # Средний объем продаж для каждого продукта
    average_sales_by_product = data.groupby('product')['sales'].mean()

    # Сводная таблица: средние продажи по регионам и продуктам
    pivot_table = data.pivot_table(values='sales', index='region', columns='product', aggfunc='mean')

    # Вывод результатов
    print("Общее количество записей:", total_records)
    print("\nРегионы в данных:", regions.tolist())
    print("\nПродукты в данных:", products.tolist())
    print("\nОбщие продажи по регионам:")
    print(total_sales_by_region)
    print("\nСредний объем продаж для каждого продукта:")
    print(average_sales_by_product)
    print("\nСводная таблица (средние продажи):")
    print(pivot_table)

except FileNotFoundError:
    print("Файл не найден. Проверьте путь к файлу.")
except pd.errors.EmptyDataError:
    print("Файл пуст. Убедитесь, что данные присутствуют.")
except Exception as e:
    print(f"Произошла ошибка: {e}")
```

**Задание 4: Получение данных из API**

Цель: Проверить способность работать с веб-данными.

**Задание:**

Используйте открытое API (например, https://jsonplaceholder.typicode.com/users).
Загрузите данные и создайте DataFrame с полями: id, name, email, company_name.
Выведите таблицу с пользователями, работающими в компаниях, название которых содержит слово "Group".
Ожидаемый ответ:

DataFrame с результатами фильтрации.

```python
import requests
import pandas as pd

# Шаг 1: Получение данных из API
url = "https://jsonplaceholder.typicode.com/users"
response = requests.get(url)
if response.status_code == 200:
    users_data = response.json()
else:
    print(f"Ошибка при получении данных из API. Код состояния: {response.status_code}")
    users_data = []

# Шаг 2: Создание DataFrame
df_users = pd.DataFrame(users_data)
df_users = df_users[['id', 'name', 'email', 'company']]

# Развернем вложенные данные компании
df_users['company_name'] = df_users['company'].apply(lambda x: x['name'] if isinstance(x, dict) else None)
df_users = df_users[['id', 'name', 'email', 'company_name']]

# Шаг 3: Фильтрация по компаниям, содержащим слово "Group"
df_filtered = df_users[df_users['company_name'].str.contains("Group", na=False)]

# Шаг 4: Вывод результата
print("Filtered Users DataFrame:")
print(df_filtered)

# Сохранение результата в файл (если требуется)
df_filtered.to_csv("filtered_users.csv", index=False)
print("\nРезультаты фильтрации сохранены в 'filtered_users.csv'.")
```
